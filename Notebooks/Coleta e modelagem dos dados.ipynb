```python
batch_size = 10000  # Tamanho fixo do lote
file_path = "../Datasets/tracks.csv"
output_file = "lyrics.csv"
```

```python
# Divide o arquivo em vários CSVs menores
def split_csv(file_path, batch_size):
    with open(file_path, 'r', encoding='utf-8') as input_file:
        reader = csv.reader(input_file)
        header = next(reader)  # Lê o cabeçalho do arquivo

        # Calcula o número total de lotes necessários com base no tamanho fixo do lote
        num_batches = math.ceil(sum(1 for _ in reader) / batch_size)
        input_file.seek(0)  # Volta para o início do arquivo

        # Divide o arquivo em vários CSVs menores
        for batch_index in range(num_batches):
            output_filename = f'batch_{batch_index + 1}.csv'
            with open(output_filename, 'w', encoding='utf-8', newline='') as output_file:
                writer = csv.writer(output_file)
                writer.writerow(header)  # Escreve o cabeçalho no arquivo de saída

                # Copia as linhas do lote atual para o arquivo de saída
                for _ in range(batch_size):
                    try:
                        writer.writerow(next(reader))
                    except StopIteration:
                        break
```

```python
import csv
import concurrent.futures

# Função que processa um arquivo CSV individualmente
def processar_arquivo_csv(numero):
    arquivo_csv_entrada = f"../Datasets/Batchs/Tracks/batch_{numero}.csv"
    arquivo_csv_saida = f"../Datasets/Batchs/Lyrics/letras_{numero}.csv"

    with open(arquivo_csv_entrada, encoding='utf-8') as arquivo_entrada, \
         open(arquivo_csv_saida, 'w', encoding='utf-8', newline='') as arquivo_saida:

        leitor_csv = csv.reader(arquivo_entrada)
        next(leitor_csv)

        escritor_csv = csv.writer(arquivo_saida)
        escritor_csv.writerow(['Artista', 'Álbum', 'Música', 'Letra'])

        for linha in leitor_csv:
            nome_artista = linha[0]
            nome_album = linha[1]
            nome_musica = linha[2]

            busca_musica = genius.search_song(nome_musica, nome_artista)

            if busca_musica is not None and busca_musica.title.lower() == nome_musica.lower() and busca_musica.artist.lower() == nome_artista.lower():
                letra_musica = busca_musica.lyrics

                # Escreve a linha no arquivo CSV de saída
                escritor_csv.writerow([nome_artista, nome_album, nome_musica, letra_musica])

# Cria um pool de trabalhadores com o número máximo de threads igual ao número de arquivos (43)
pool = concurrent.futures.ThreadPoolExecutor(max_workers=43)

# Executa as tarefas em paralelo para os arquivos CSV
for i in range(1, 44):
    pool.submit(processar_arquivo_csv, i)

# Aguarda a conclusão de todas as tarefas
pool.shutdown()
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time

# Função para obter os IDs das faixas em lote
def get_track_ids(track_names, artist_names):
    # Defina suas credenciais do Spotify (obtenha-as em https://developer.spotify.com/dashboard/)
    client_id = ''
    client_secret = ''

    # Configuração da autenticação
    client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

    # Lista para armazenar os IDs das faixas
    track_ids = []

    # Tamanho máximo de busca em lote é 50
    batch_size = 50

    # Dividir as faixas em lotes menores
    batches = [track_names[i:i + batch_size] for i in range(0, len(track_names), batch_size)]

    # Iterar sobre cada lote
    for batch in batches:
        # Realizar a busca em lote
        results = sp.search(q=batch, type='track', limit=batch_size)

        # Percorrer os resultados e obter os IDs das faixas
        for track_result in results['tracks']['items']:
            track_ids.append(track_result['id'])

        # Aguardar 1 segundo para evitar exceder os limites da API do Spotify
        time.sleep(1)

    return track_ids

# Caminho do arquivo de entrada e do arquivo de saída
input_file = "../Datasets/lyrics.csv"
output_file = "../Datasets/tracks_w_attributes_lyrics.csv"

# Criar o arquivo de saída com as colunas necessárias
output_columns = ['Artist', 'Album', 'Track', 'Lyrics', 'track_id']
output_df = pd.DataFrame(columns=output_columns)
output_df.to_csv(output_file, index=False)

# Carregar o arquivo CSV com as informações das faixas em lotes
batch_size = 100
for df_chunk in pd.read_csv(input_file, chunksize=batch_size):
    # Obter os nomes das faixas e artistas do DataFrame
    track_names = df_chunk['Track'].tolist()
    artist_names = df_chunk['Artist'].tolist()

    # Obter os IDs das faixas em lote
    track_ids = get_track_ids(track_names, artist_names)

    # Adicionar a coluna 'track_id_temp' ao DataFrame
    df_chunk['track_id_temp'] = pd.Series(track_ids)

    # Filtrar as linhas onde o ID da faixa foi encontrado
    df_filtered = df_chunk.dropna(subset=['track_id_temp'])

    # Atualizar apenas as linhas correspondentes aos IDs filtrados
    df_chunk.loc[df_filtered.index, 'track_id'] = df_filtered['track_id_temp']

    # Adicionar as linhas ao arquivo de saída
    df_chunk.to_csv(output_file, mode='a', header=False, index=False)

    # Imprimir informações sobre o progresso
    last_row_index = df_chunk.index[-1]
    print(f"Processando até a linha {last_row_index} do arquivo de origem.")

print("Processamento concluído.")
```

## Rotulando língua das letras entre inglês ou português

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

# Baixando os recursos necessários do NLTK
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Definindo uma lista de palavras vazias (stop words) em português
stop_words_pt = set(stopwords.words('portuguese'))

# Função para verificar se uma palavra está em inglês
def is_english_word(word):
    return word.lower() not in stop_words_pt

# Função para rotular as letras como inglês ou português
def label_language(row):
    lyrics = row['Lyrics']
    if isinstance(lyrics, str):
        words = word_tokenize(lyrics)  # Tokeniza as palavras
        english_count = sum(1 for word in words if is_english_word(word))
        portuguese_count = len(words) - english_count

        if english_count > portuguese_count:
            return 'English'
        else:
            return 'Portuguese'
    else:
        return 'Unknown'

# Importe o arquivo CSV para um DataFrame
df = pd.read_csv('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/lyrics_fixed.csv')
total_rows = df.shape[0]  # Número total de linhas

# Variáveis para acompanhar o progresso
processed_rows = 0
remaining_rows = total_rows

# Aplica a função 'label_language' a cada linha do DataFrame e cria uma nova coluna 'Language'
for index, row in df.iterrows():
    processed_rows += 1
    remaining_rows -= 1

    df.at[index, 'Language'] = label_language(row)

    # Imprime o progresso
    print(f"Processando linha {processed_rows} de {total_rows}. Faltam {remaining_rows} linhas.")

# Salva o DataFrame modificado em um novo arquivo CSV
df.to_csv('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/lyrics_w_language.csv', index=False)
```

## Limpeza

```python
import pandas as pd
import re

def limpar_letras(letra):
    if pd.isnull(letra):
        return letra

    # Remove as palavras indesejadas
    indesejadas = ['intro', 'verse', 'espaol', 'pre-chorus', 'chorus', 'bridge', 'pre', 'pós', 'lyrics', 'embed', 'contributors', 'translations', 'pré', 'refrão', 'pré-refrão', 'verso', 'ponte', 'entrada', 'saída', 'english', 'español', 'português', '中文', 'हिन्दी', 'русский', 'العربية', '日本語', 'deutsch', 'français', '한국어', 'italiano', 'türkçe', 'polski', 'bahasa indonesia', 'اردو', 'nederlands', 'català', 'svenska', 'čeština', 'magyar', 'norsk', 'suomi', 'dansk', 'română', 'slovenčina', 'slovenščina', 'tiếng việt', 'українська', 'ελληνικά', 'ภาษาไทย', 'עברית', 'हिन्दी', 'தமிழ்', 'filipino', 'bahasa melayu', 'kiswahili', 'ភាសាខ្មែរ', 'қазақша', 'português (brasil)', 'українська', 'മലയാളം', 'slovenščina', 'cymraeg', 'euskara', 'gaeilge', 'kinyarwanda', 'हिन्दी', 'esperanto', 'ગુજરાતી', 'sardu', 'ગુજરાતી', 'ਪੰਜਾਬੀ', 'हिन्दी', 'cymraeg', 'sardu', 'kiswahili', 'català', 'lëtzebuergesch', 'kurdî', 'ਪੰਜਾਬੀ', 'cymraeg', '한국어', 'ଓଡ଼ିଆ', 'मराठी', 'cymraeg', 'हिन्दी', 'esperanto', 'català', 'ગુજરાતી', 'sardu', 'ਪੰਜਾਬੀ', 'euskara', 'ગુજરા']

    # Remove as palavras indesejadas com regex
    regex = re.compile(r'\b(?:{})\b'.format('|'.join(indesejadas)), flags=re.IGNORECASE)
    letra = regex.sub('', str(letra))

    # # Remove palavras compostas e separadas por camel case
    # regex = re.compile(r'(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])')
    # letra = regex.sub(' ', letra)

    # Remove números
    letra = re.sub(r'\d+', '', letra)

    # Remove espaços extras resultantes
    letra = re.sub('\s+', ' ', letra).strip()

    return letra

# Carrega o CSV original
lyrics_clean = pd.read_csv('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/lyrics_formatado_7.csv')

# Aplica a função limpar_letras à coluna 'Lyrics' do CSV
lyrics_clean['Lyrics'] = lyrics_clean['Lyrics'].apply(limpar_letras)

# Salva os dados atualizados em um novo CSV
lyrics_clean.to_csv('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/lyrics_formatado_8.csv', index=False)
```

```python
import pandas as pd
import glob

# Obtenha uma lista de todos os arquivos CSV no diretório atual
arquivos_csv = glob.glob('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/combinar/*.csv')

# Crie uma lista vazia para armazenar os DataFrames
dataframes = []

# Percorra cada arquivo CSV e carregue-o como um DataFrame, ignorando o header
for arquivo in arquivos_csv:
    df = pd.read_csv(arquivo, usecols=["Artist", "Song", "Album", "Genres", "Parent Genres", "Album Date", "Time", "Danceability", "Energy", "Acousticness", "Instrumentalness", "Happiness", "Speechiness", "Liveness", "Loudness", "Spotify Track Id"])
    dataframes.append(df)

# Concatene os DataFrames em um único DataFrame
df_final = pd.concat(dataframes, ignore_index=True)

# Salve o DataFrame final em um novo arquivo CSV
df_final.to_csv('/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/faixas_w_a.csv', index=False)
```

```python
from pyspark.sql import SparkSession

# Inicializa a sessão do Spark
spark = SparkSession.builder.getOrCreate()

# Carrega os dados dos scrobbles
scrobbles_data = spark.read.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/1685893158_r.csv", header=True)

# Carrega os dados das faixas
faixas_data = spark.read.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/faixas_w_a.csv", header=True)

# Seleciona as colunas relevantes dos scrobbles
scrobbles_data = scrobbles_data.select("uts", "utc_time", "artist", "album", "track", "user_id")

# Seleciona as colunas relevantes das faixas
faixas_data = faixas_data.select("Song", "Artist", "Genres", "Parent Genres", "Album Date", "Time", "Danceability",
                                 "Energy", "Acousticness", "Instrumentalness", "Happiness", "Speechiness",
                                 "Liveness", "Loudness")

# Renomeia as colunas das faixas para corresponder às colunas dos scrobbles
faixas_data = faixas_data.withColumnRenamed("Song", "track").withColumnRenamed("Artist", "artist")

# Combina os dados dos scrobbles e das faixas com base nas colunas "artist" e "track"
combined_data = scrobbles_data.join(faixas_data, on=["artist", "track"], how="inner")

# Salva o resultado em um novo arquivo CSV
combined_data.write.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/scrobbles_1685893158_w_a.csv", header=True, mode="overwrite")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import when

# Inicializa a sessão do Spark
spark = SparkSession.builder.getOrCreate()

# Carrega os dados dos scrobbles
scrobbles_data = spark.read.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/1685893158_r.csv", header=True)

# Carrega os dados das faixas
faixas_data = spark.read.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/faixas_w_a.csv", header=True)

# Seleciona as colunas relevantes dos scrobbles
scrobbles_data = scrobbles_data.select("uts", "utc_time", "artist", "album", "track", "user_id")

# Seleciona as colunas relevantes das faixas
faixas_data = faixas_data.select("Song", "Artist", "Genres", "Parent Genres", "Album Date", "Time", "Danceability",
                                 "Energy", "Acousticness", "Instrumentalness", "Happiness", "Speechiness",
                                 "Liveness", "Loudness")

# Renomeia as colunas das faixas para corresponder às colunas dos scrobbles
faixas_data = faixas_data.withColumnRenamed("Song", "track").withColumnRenamed("Artist", "artist")

# Combina os dados dos scrobbles e das faixas com base nas colunas "artist" e "track"
scrobbles = scrobbles_data.join(faixas_data, on=["artist", "track"], how="inner")

# Adiciona a coluna "mood" baseada na coluna "Happiness"
scrobbles = scrobbles.withColumn("mood", when(scrobbles["Happiness"] >= 56.0, 1).otherwise(0))

# Salva o resultado em um novo arquivo CSV
scrobbles.write.csv("/content/drive/MyDrive/TCC/KledysonHenrique/Datasets/Batchs/temp/scrobbles_1685893158_w_a.csv", header=True, mode="overwrite")
```
